```python
import gym
import numpy as np
import time

env = gym.make('FrozenLake-v1',render_mode='human')

STATES = env.observation_space.n # Number of states
ACTIONS = env.action_space.n # Number of actions

Q = np.zeros((STATES,ACTIONS)) # CREATE a matrix with all 0 values

EPISODES = 1500 # EPOCHS
MAX_STEPS = 100 # STEPS PER EPISOD

LEARNING_RATE = 0.81
GAMMA = 0.96

RENDER = False

epsilon = 0.9 # Start with 90% chance to pick a random action
```

```python
rewards = []
RENDER = False
s = "#"
m = "="

for episode in range(EPISODES):
  i = episode
  print("STEPS : "+ i*m + ' >'+s+'       '+ str(i+1)+'/'+ str(EPISODES))
  state = env.reset()

  for _ in range(MAX_STEPS):
    
    if RENDER:
      env.render()

    if np.random.uniform(0,1) < epsilon :
      action = env.action_space.sample() # take a random action

    else :
      action = np.argmax(Q[state, : ]) # Use Q-Table to pick the best action based on the curent state
    
    next_state ,reward,done , _ = env.step(action)

    Q[state,action] = Q[state,action] + LEARNING_RATE * (reward + GAMMA * np.max(Q[next_state, : ]) - Q[state,action])

    state = next_state

    if done :
      rewards.append(reward)

      epsilon -= 0.001

      break

print(Q)


print(f"AVERAGE REWARD : {sum(rewards)/len(rewards)}")
```
